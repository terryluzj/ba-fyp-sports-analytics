```r
data <- read.csv("regression_data.csv", encoding="UTF-8")
data
```

## Residual ~ Age

```r
date <- data$Date
time <- data$Time
name <- data$Name
data[, c("Trainer", "X", "No..of.Contestants", "Date", "Time", "Name")] <- NULL
ols <- lm(Time.Residual ~ as.factor(Age), data=data)
summary(ols)
```

## Time ~ Age

```r
ols <- lm(time ~ as.factor(Age), data=data)
summary(ols)
```

## Residual ~ .

```r
data$Age <- as.factor(data$Age)
data$Track.No. <- as.factor(data$Track.No.)
data$Race.Number <- as.factor(data$Race.Number)
ols <- lm(Time.Residual ~ ., data=data)
summary(ols)
```

## Time ~ .

```r
ols <- lm(time ~ ., data=data)
summary(ols)
```

## Panel Model

```r
library(plm)
data$Date = date
data$Name = name
data$Time = time
data
```

```r
fixed <- plm(Time ~ Age, data=data, index=c("Date", "Name"), model="within")
summary(fixed)
```

```r
fixed <- plm(Time.Residual ~ Age, data=data, index=c("Date", "Name"), model="within")
summary(fixed)
```

```r
fixed <- plm(Time ~ Age + Weight, data=data, index=c("Date", "Name"), model="within")
summary(fixed)
```
